### Matplotlib_for_3D_visualization

**Plot 3D graph for multiple columns using matplotlib**

```python
# import the library and also the dataset
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ADANIPORTS.csv', parse_dates = True)
```

```python
# this is the head of the dataset
df.head()
```

# let's calculate High and low


```python
df['H-L'] = df.High - df.Low
df['100MA'] = df['Close'].rolling(100).mean()
```

# Create 3D graph


```python
# Create axes
# we are mentioning matplotlib that the projection is three deals
ax = plt.axes(projection = '3d')
ax.scatter(df.index,df['H-L'], df['100MA'])

# set labels
ax.set_xlabel('Index')
ax.set_ylabel('H-L')
ax.set_zlabel('100MA')
plt.show()
```

As you can see we know have a 3D plot of index minus low and 100 moving average.
let's try doing the same other columns.

let's do it with volume and see if there is any correlaion

```python
# Create axes
# we are mentioning matplotlib that the projection is three deals
ax = plt.axes(projection = '3d')
ax.scatter(df.index,df['H-L'], df['Volume'])

# set labels
ax.set_xlabel('Index')
ax.set_ylabel('H-L')
ax.set_zlabel('Volume')
plt.show()
```

Here is the 3D graph. And you can see, there  is the correlation between high and lows and volume.
