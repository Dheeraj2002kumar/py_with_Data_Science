```python
import pandas as pd
```

```python
df = pd.read_csv('HousePrices.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
# plot the graph
df.plot()
```

```python
df['price'].plot()
```

```python
# print a legend that explain the column name of the graph is plotted
df['price'].plot(legend = True)
```

```python
# plot the scatter plot
df.plot(x = 'price', y = 'bedrooms', kind = 'scatter', legend = True)
```

```python
# we can also add title and modify the figersize using the title & figersize parameter
df.plot(title = 'Sample Plot', figsize = (15, 10))
```

```python
# we can also plot box plot & Histogram by passing parameter in kind
df['price'].plot(kind='box', title = 'Box Plot')
```

```python
# we can also change in histogram
df['price'].plot(kind='hist', title = 'Histogram Plot')
```
