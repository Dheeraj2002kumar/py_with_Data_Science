### Steps to Create Plot - Example


### Step 01
Import the required libraries

```python
# import numpy for generating random numbers
import numpy as np

#import matplotlib library
# pyplot plot the numbers
import matplotlib.pyplot as plt

# style classes use set the grid style
from matplotlib import style


# %matplotlib inline disply to required plot in jupyter notebook
%matplotlib inline
```

### Step 02
Define or import the required dataset

```python
# generate random numbers (total 10)
#Defined the dataset ==>  used NumPy random method

# note that the range is 10
randomNumber = np.random.rand(10)
```

```python
# print method ==> view the created random numbers
# view them
print(randomNumber)
```

### Step 03
Set the plot parameters

```python
# select the style of the plot
# ggplot ==> set the style of the plot
style.use('ggplot')

# plot the random number
# 'g' denote line color is green
# lebel denote ==> set the legend lebel
# linewidth ==> set line width 2
plt.plot(randomNumber, 'g', label = 'line one', linewidth = 2)

# set coordinates labels

#x axis is number of random numbers (index)
plt.xlabel('Range')

# y axis is actual random number
plt.ylabel('Numbers')

# Title of the plot
# set the title 'First Plot'
plt.title('First Plot')
```

### Step 04
Display the create plot

```python
# Plot the graph
plt.legend('line one')

# show method Display the created plot
plt.show()
```
