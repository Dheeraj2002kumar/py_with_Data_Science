### <h1><b>Problem Statement</b></h1>
<h3>You have been provided with a dataset that lists Ohio State's leading causes of death from the year 2012.</h3>
<h3>Usind the two data points cause of death and percentile, draw a pie chart to visualize the dataset.</h3>

```python
# import required libraries
import matplotlib.pyplot as plt

# %matplotlib inline ==> is use for view the plot in jupyter notebook
%matplotlib inline
```

```python
# Data from Ohio state, Leading causes of death in the state from year 2012
# cause of deaths
cause = 'Chronic Diseases', 'Unintentional Injuries', 'Alzheimers','Infuenze and Pneumonia', 'Sepsis', 'Others'

# percentile
percentile = [62, 5, 4, 2,1,26]
```

```python
# set the pie chart plot properties 
# set figure size 
plt.figure(figsize=(10,10))

# explode the largest pie in the data set 
explode = (0.05,0,0,0,0,0)

# set pie chart properties 
plt.pie(percentile, labels = cause, explode = explode, autopct = '%1.1f%%', startangle = 70)

# set axis equal to drow  pie as circle
plt.axis('equal')

# set title of the pie chart
plt.title('Oho state - 2012 : Leading Causes of Death')

# show the plot (pie chart)
plt.show(12)
```
