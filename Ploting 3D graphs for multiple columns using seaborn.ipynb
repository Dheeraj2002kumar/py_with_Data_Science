### Ploting 3D graphs for multiple columns using seaborn 

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ADANIPORTS.csv', parse_dates=True)
```

“parse_dates=True” in Pandas refers to automatically parsing specified columns or the index of a DataFrame as datetime objects during the read_csv() operation.

When using pandas.read_csv(), you can set “parse_dates=True” to parse the index as datetime. If no column is specified, pandas will attempt to parse the first few columns as dates

```python
df['H-L'] =  df.High - df.Low
```

```python
df['100MA'] = df['Close'].rolling(100).mean()
```

```python
# set style darkgrid
sns.set_style('darkgrid')
```

```python
# Ploting 3D graph
ax = plt.axes(projection = '3d')
ax.scatter(df.index,df['H-L'], df['100MA'])

# set labels
ax.set_xlabel('Index')
ax.set_ylabel('H-L')
ax.set_zlabel('100MA')
plt.show()
```

As you can see, this is a 3D graph. But with dark grid.

**let's look at other graph and other method that we can use to plot 3D graphs**

```python
import numpy as np
```

```python
z1 = np.linspace(0, 10, 100)
x1 = np.cos(2*z1)
y1 = np.sin(2*z1)
```

==> Let's set the style two, a white grid and plot the graph

```python
sns.set_style('whitegrid')
ax = plt.axes(projection = '3d')

# we will using 3D method
ax.plot3D(x1, y1, z1)
plt.show()
```

we can see the Spiral in 3D graph. We can increase the spiral by increasing the value of X one

```python
z1 = np.linspace(0, 10, 100)
x1 = np.cos(4*z1)
y1 = np.sin(4*z1)
```

```python
sns.set_style('whitegrid')
ax = plt.axes(projection = '3d')

# we will using 3D method
ax.plot3D(x1, y1, z1)
plt.show()
```

As you can see the spiral has increased

**Let's explore the 3D surface and try to plot it using Seaborn.**

Let's define the data and functions that are needed to plot the mash grid

```python
def return_z(x,y):
    return 50-(x**2+y**2)
```

```python
sns.set_style('whitegrid')
```

```python
x1,y1 = np.linspace(-5,5,50),np.linspace(-5,5,50)
```

```python
#Let's create a mash grid using numpy
x1,y1 = np.meshgrid(x1,y1)
z1 = return_z(x1,y1)
```

we now have the data. So let's go ahead and plot it

```python
# Remember to use projection as 3D
ax = plt.axes(projection = '3d')

# use the plot surface method
ax.plot_surface(x1,y1,z1)

plt.show()
```

As you can see, A mesh grid has been plotted.

These are the some of the methods and ways that we can plot 3D graphs using seaborn.
