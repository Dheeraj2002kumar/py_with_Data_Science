### Using_matplotlib_to_plot_graph

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('HousePrices.csv', parse_dates = ['date'], index_col = ['date'])
df.sort_index(inplace=True)
```

```python
df.head()
```

```python
# line plot for the above data
plt.plot(df['price'])
```

```python
# boxplot for the same data
# plot a boxplot
plt.boxplot(df['price'])
```

```python
# histogram for the same data
# plot a histogram
plt.hist(df['sqft_living'])
```

# plot the scatterplot and also explore the other feature that matplotlib has to offer

```python
# set the labels
plt.xlabel(['price'])
plt.ylabel(['sqft_living'])

# set the title
plt.title('Scatter plot')

plt.scatter(x = df['price'], y = df['sqft_living'])
```

# create a simple dictionary to view bar and  pie plot 

```python
d = {'a': 10, 'b': 20, 'c': 30}
```

```python
# bar plot for the same data
plt.bar(x = d.keys(), height = d.values())
```

```python
# pie plot for the same data
plt.pie(x = d.values(), labels = d.keys())
```

# we can also manipulate the feature of a figure for a line plot


```python
plt.plot(df['price'])
```

# modify of a properties of a figure and save it


```python
# dpi ==> display pixels per inch will increase the resolution of a graph
plt.figure(figsize=(15,10), dpi = 100)
plt.xlabel('Time')
plt.ylabel('Price')
plt.title('line Plot of Price')
plt.legend()
plt.plot(df['price'])
```

# let's save it

```python
# dpi ==> display pixels per inch will increase the resolution of a graph
plt.figure(figsize=(15,10), dpi = 100)
plt.xlabel('Time')
plt.ylabel('Price')
plt.title('line Plot of Price')
plt.plot(df['price'])
plt.savefig('Price.png')
```

# when you look our local directory, you will be able to see the exact same graph

These are the some feature that matplotlib offer for visualizing the data
